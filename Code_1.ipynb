```python
import numpy as np

np.random.seed(42)

def choose_next_point_simple(X, y):
    dim = X.shape[1]
    
    # znajdź najlepszy punkt
    idx_best = np.argmax(y)
    x_best = X[idx_best]
    
    # 70% exploit, 30% explore
    if np.random.rand() < 0.3:
        # eksploracja (losowy punkt)
        return np.random.uniform(0, 1, dim)
    
    # eksploatacja (blisko najlepszego)
    noise = np.random.normal(0, 0.05, dim)
    x_new = x_best + noise
    
    return np.clip(x_new, 0, 1)


def format_query(x):
    return "-".join([f"{xi:.6f}" for xi in x])
```

```python
# Function 1
q1 = choose_next_point_simple(initial_inputs_1, initial_outputs_1)
print("F1:", format_query(q1))

# Function 2
q2 = choose_next_point_simple(initial_inputs_2, initial_outputs_2)
print("F2:", format_query(q2))

# Function 3
q3 = choose_next_point_simple(initial_inputs_3, initial_outputs_3)
print("F3:", format_query(q3))

# Function 4
q4 = choose_next_point_simple(initial_inputs_4, initial_outputs_4)
print("F4:", format_query(q4))

# Function 5
q5 = choose_next_point_simple(initial_inputs_5, initial_outputs_5)
print("F5:", format_query(q5))

# Function 6
q6 = choose_next_point_simple(initial_inputs_6, initial_outputs_6)
print("F6:", format_query(q6))

# Function 7
q7 = choose_next_point_simple(initial_inputs_7, initial_outputs_7)
print("F7:", format_query(q7))

# Function 8
q8 = choose_next_point_simple(initial_inputs_8, initial_outputs_8)
print("F8:", format_query(q8))
```
