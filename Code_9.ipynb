```python
# ======================
# RESTRUCTURE ITERATION DATA
# ======================

n_weeks = len(inputs_raw)
n_functions = 8

iter_inputs = []
iter_outputs = []

for f in range(n_functions):
    X_f = []
    y_f = []

    for w in range(n_weeks):
        x = np.array(inputs_raw[w][f])
        y = float(outputs_raw[w][f])

        X_f.append(x)
        y_f.append(y)

    X_f = np.vstack(X_f)
    y_f = np.array(y_f)

    iter_inputs.append(X_f)
    iter_outputs.append(y_f)


# ======================
# COLLECT INITIAL DATA INTO LISTS
# ======================

initial_inputs = [
    initial_inputs_1,
    initial_inputs_2,
    initial_inputs_3,
    initial_inputs_4,
    initial_inputs_5,
    initial_inputs_6,
    initial_inputs_7,
    initial_inputs_8,
]

initial_outputs = [
    initial_outputs_1,
    initial_outputs_2,
    initial_outputs_3,
    initial_outputs_4,
    initial_outputs_5,
    initial_outputs_6,
    initial_outputs_7,
    initial_outputs_8,
]


# ======================
# MERGE ALL DATA
# ======================

X_final = []
y_final = []

for i in range(8):

    X_combined = np.vstack([
        initial_inputs[i],
        iter_inputs[i]
    ])

    y_combined = np.concatenate([
        initial_outputs[i],
        iter_outputs[i]
    ])

    X_final.append(X_combined)
    y_final.append(y_combined)

    print(f"Function {i+1}: {X_combined.shape}, {y_combined.shape}")


# ======================
# READY FOR MODEL
# ======================

# Example usage:
# for i in range(8):
#     q = choose_next_point_v7(X_final[i], y_final[i])
#     print(f"Function {i+1}: " + "-".join([f"{v:.6f}" for v in q]))
```

```python
import numpy as np
import tensorflow as tf

# ======================
# MODEL (MC DROPOUT + STABILIZACJA)
# ======================

def build_model(dim, lr=0.002):
    model = tf.keras.Sequential([
        tf.keras.Input(shape=(dim,)),
        tf.keras.layers.Dense(32, activation='relu'),
        tf.keras.layers.Dropout(0.25),
        tf.keras.layers.Dense(32, activation='relu'),
        tf.keras.layers.Dropout(0.25),
        tf.keras.layers.Dense(1)
    ])
    model.compile(optimizer=tf.keras.optimizers.Adam(lr), loss='mse')
    return model


def mc_predict(model, x, T=12):
    preds = []
    for _ in range(T):
        preds.append(model(x, training=True).numpy().squeeze())
    preds = np.array(preds)
    return preds.mean(), preds.std()


# ======================
# TRAIN MODEL
# ======================

def train_model(X, y):
    model = build_model(X.shape[1])
    model.fit(X, y, epochs=160, verbose=0)
    return model


# ======================
# TRUST REGION + CEM GENERATION
# ======================

def generate_candidates(X, y, dim, n=30):

    # elite selection
    k = max(4, len(X)//4)
    elite_idx = np.argsort(y)[-k:]
    elite = X[elite_idx]

    # centroid of best region
    center = elite.mean(axis=0)

    # stable variance (prevents collapse)
    spread = elite.std(axis=0)
    spread = np.maximum(spread, 0.08)

    candidates = []

    for _ in range(n):

        # 70% trust region sampling
        if np.random.rand() < 0.7:
            x = np.random.normal(center, spread)

        # 30% global exploration
        else:
            x = np.random.uniform(0, 1, dim)

        # small jitter (prevents symmetry collapse)
        x += np.random.normal(0, 0.02, dim)

        candidates.append(np.clip(x, 0, 1))

    return np.array(candidates)


# ======================
# SCORING FUNCTION (FIXES F8 COLLAPSE)
# ======================

def score_point(model, x, global_std):

    mu, sigma = mc_predict(model, x.reshape(1, -1))

    kappa = 0.15 + 0.35 * (global_std / (np.abs(mu) + 1e-8))

    # anti-boundary penalty (CRITICAL FIX)
    boundary_penalty = np.mean(np.minimum(x, 1 - x))

    return mu + kappa * sigma + 0.25 * boundary_penalty


# ======================
# MAIN OPTIMISATION STEP
# ======================

def choose_next(X, y):

    dim = X.shape[1]

    # normalize (stability)
    y_norm = (y - y.mean()) / (y.std() + 1e-8)

    model = train_model(X, y_norm)

    candidates = generate_candidates(X, y, dim)

    global_std = np.std(y)

    best_x = None
    best_score = -1e9

    for x in candidates:

        # extra safety: occasional full exploration override
        if np.random.rand() < 0.15:
            x = np.random.uniform(0, 1, dim)

        s = score_point(model, x, global_std)

        if s > best_score:
            best_score = s
            best_x = x

    return best_x


# ======================
# GENERATE ALL QUERIES
# ======================

queries = []

for i in range(8):
    X = X_final[i]
    y = y_final[i]

    q = choose_next(X, y)
    queries.append(q)


# ======================
# OUTPUT FORMAT
# ======================

for i, q in enumerate(queries):
    print(f"Function {i+1}: " + "-".join([f"{v:.6f}" for v in q]))
```
