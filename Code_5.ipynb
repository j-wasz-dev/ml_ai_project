```python
import numpy as np
import tensorflow as tf

np.random.seed(42)
tf.random.set_seed(42)

# ----------------------
# Build model (slightly regularised)
# ----------------------
def build_model(dim):
    model = tf.keras.Sequential([
        tf.keras.layers.Dense(32, activation='relu', input_shape=(dim,)),
        tf.keras.layers.Dropout(0.1),
        tf.keras.layers.Dense(32, activation='relu'),
        tf.keras.layers.Dense(1)
    ])

    model.compile(
        optimizer=tf.keras.optimizers.Adam(learning_rate=0.01),
        loss='mse'
    )
    return model


# ----------------------
# Train ensemble
# ----------------------
def train_ensemble(X, y, n_models=5):
    models = []
    for _ in range(n_models):
        model = build_model(X.shape[1])
        model.fit(X, y, epochs=120, verbose=0)
        models.append(model)
    return models


# ----------------------
# Ensemble prediction
# ----------------------
def ensemble_predict(models, x):
    preds = np.array([m.predict(x, verbose=0)[0, 0] for m in models])
    return preds.mean(), preds.std()


# ----------------------
# Gradient ascent using ONE model
# ----------------------
def gradient_ascent(model, x_init, steps=40, lr=0.01):
    x = tf.Variable(x_init.reshape(1, -1), dtype=tf.float32)

    for _ in range(steps):
        with tf.GradientTape() as tape:
            tape.watch(x)
            pred = model(x)

        grads = tape.gradient(pred, x)

        # clip gradients (stability)
        grads = tf.clip_by_value(grads, -0.1, 0.1)

        x.assign_add(lr * grads)
        x.assign(tf.clip_by_value(x, 0.0, 1.0))

    return x.numpy().flatten()


# ----------------------
# MAIN STRATEGY
# ----------------------
def choose_next_point_ensemble(X, y):

    dim = X.shape[1]

    # ----------------------
    # 1. Train ensemble
    # ----------------------
    models = train_ensemble(X, y, n_models=5)

    # ----------------------
    # 2. Generate candidates
    # ----------------------
    candidates = []

    best_x = X[np.argmax(y)]

    # (A) Gradient-based candidates
    for m in models:
        x0 = best_x if np.random.rand() < 0.7 else np.random.uniform(0, 1, dim)
        x_new = gradient_ascent(m, x0)
        candidates.append(x_new)

    # (B) Random exploration
    for _ in range(5):
        candidates.append(np.random.uniform(0, 1, dim))

    # (C) Local perturbations
    for _ in range(5):
        noise = np.random.normal(0, 0.05, dim)
        x_new = np.clip(best_x + noise, 0, 1)
        candidates.append(x_new)

    # ----------------------
    # 3. Score candidates
    # ----------------------
    best_score = -np.inf
    best_candidate = None

    for x in candidates:
        x_tf = x.reshape(1, -1)

        mu, sigma = ensemble_predict(models, x_tf)

        # gradient norm (importance)
        with tf.GradientTape() as tape:
            xt = tf.Variable(x_tf, dtype=tf.float32)
            tape.watch(xt)
            pred = models[0](xt)

        grad = tape.gradient(pred, xt).numpy()
        grad_norm = np.linalg.norm(grad)

        # FINAL SCORE
        score = mu + 0.2 * sigma + 0.05 * grad_norm

        if score > best_score:
            best_score = score
            best_candidate = x

    return best_candidate


# ----------------------
# FORMAT
# ----------------------
def format_query(x):
    return "-".join([f"{xi:.6f}" for xi in x])


# ----------------------
# RUN FOR ALL FUNCTIONS
# ----------------------
queries = []

for i in range(8):
    X = inputs_15[i]
    y = outputs_15[i]

    q = choose_next_point_ensemble(X, y)
    queries.append(q)


# ----------------------
# PRINT
# ----------------------
for i, q in enumerate(queries):
    print(f"Function {i+1}: {format_query(q)}")
```
