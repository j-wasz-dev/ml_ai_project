```python
import numpy as np
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import Matern
from sklearn.svm import SVC

np.random.seed(42)

def choose_next_point_hybrid(X, y, n_candidates=3000):
    dim = X.shape[1]

    # ----------------------
    # 1. Train GP
    # ----------------------
    
    kernel = Matern(length_scale=0.3, nu=1.5)

    gp = GaussianProcessRegressor(
        kernel=kernel,
        alpha=1e-1,
        normalize_y=True,
        optimizer=None   # 🔥 KEY: disables problematic optimisation
    )
    
    gp.fit(X, y)

    # ----------------------
    # 2. Generate candidates
    # ----------------------
    X_candidates = np.random.uniform(0, 1, (n_candidates, dim))

    # ----------------------
    # 3. GP prediction
    # ----------------------
    mu, sigma = gp.predict(X_candidates, return_std=True)

    # ----------------------
    # 4. SVM (safe version)
    # ----------------------
    threshold = np.median(y)
    labels = (y >= threshold).astype(int)

    if len(np.unique(labels)) < 2:
        prob_good = np.ones(n_candidates) * 0.5
    else:
        svm = SVC(kernel='rbf', probability=True)
        svm.fit(X, labels)
        prob_good = svm.predict_proba(X_candidates)[:, 1]

    # ----------------------
    # 5. Combined score
    # ----------------------
    score = mu + 0.15 * sigma + 0.15 * prob_good

    return X_candidates[np.argmax(score)]


def format_query(x):
    return "-".join([f"{xi:.6f}" for xi in x])
```
