```python
import numpy as np
import tensorflow as tf

np.random.seed(42)
tf.random.set_seed(42)

# ----------------------
# Build model
# ----------------------
def build_model(dim):
    model = tf.keras.Sequential([
        tf.keras.layers.Dense(32, activation='relu', input_shape=(dim,)),
        tf.keras.layers.Dense(32, activation='relu'),
        tf.keras.layers.Dense(1)
    ])
    
    model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.01),
                  loss='mse')
    return model


def choose_next_point_nn(X, y, n_steps=50, lr=0.01):
    dim = X.shape[1]

    # ----------------------
    # 1. Train NN
    # ----------------------
    model = build_model(dim)
    model.fit(X, y, epochs=150, verbose=0)

    # ----------------------
    # 2. Smart starting point
    # ----------------------
    if np.random.rand() < 0.5:
        x = tf.Variable(X[np.argmax(y)].reshape(1, -1), dtype=tf.float32)
    else:
        x = tf.Variable(np.random.uniform(0, 1, (1, dim)), dtype=tf.float32)

    # ----------------------
    # 3. Gradient ascent
    # ----------------------
    for _ in range(n_steps):
        with tf.GradientTape() as tape:
            tape.watch(x)
            pred = model(x)

        grads = tape.gradient(pred, x)

        # gradient clipping
        step = tf.clip_by_value(grads, -0.1, 0.1)
        x.assign_add(lr * step)

        # keep in bounds
        x.assign(tf.clip_by_value(x, 0.0, 1.0))

    x_final = x.numpy().flatten()

    # ----------------------
    # 4. ADD SMALL NOISE (NEW)
    # ----------------------
    noise = np.random.normal(0, 0.01, size=dim)
    x_final = np.clip(x_final + noise, 0, 1)

    # ----------------------
    # 5. RANDOM EXPLORATION
    # ----------------------
    if np.random.rand() < 0.3:
        return np.random.uniform(0, 1, dim)

    return x_final


def format_query(x):
    return "-".join([f"{xi:.6f}" for xi in x])
```

```python
queries = []

for i in range(8):
    X = inputs[i]
    y = outputs[i]

    q = choose_next_point_nn(X, y)
    queries.append(q)
```

```python
for i, q in enumerate(queries):
    formatted = "-".join([f"{x:.6f}" for x in q])
    print(f"Function {i+1}: {formatted}")
```
