```python
####IMPORTING
import re
import numpy as np

with open("data_7/inputs.txt", "r") as f:
    text = f.read()

blocks = re.findall(r'\[.*?\](?=\s*\[|$)', text, re.DOTALL)
inputs_raw = [eval(b, {"array": np.array}) for b in blocks]


with open("data_7/outputs.txt", "r") as f:
    text = f.read()

blocks = re.findall(r'\[.*?\](?=\s*\[|$)', text, re.DOTALL)
outputs_raw = [eval(b, {"np": np}) for b in blocks]

print(inputs_raw)
print(outputs_raw)
```

```python
# ======================
# RESTRUCTURE ITERATION DATA
# ======================

n_weeks = len(inputs_raw)
n_functions = 8

iter_inputs = []
iter_outputs = []

for f in range(n_functions):
    X_f = []
    y_f = []

    for w in range(n_weeks):
        x = np.array(inputs_raw[w][f])
        y = float(outputs_raw[w][f])

        X_f.append(x)
        y_f.append(y)

    X_f = np.vstack(X_f)
    y_f = np.array(y_f)

    iter_inputs.append(X_f)
    iter_outputs.append(y_f)


# ======================
# COLLECT INITIAL DATA INTO LISTS
# ======================

initial_inputs = [
    initial_inputs_1,
    initial_inputs_2,
    initial_inputs_3,
    initial_inputs_4,
    initial_inputs_5,
    initial_inputs_6,
    initial_inputs_7,
    initial_inputs_8,
]

initial_outputs = [
    initial_outputs_1,
    initial_outputs_2,
    initial_outputs_3,
    initial_outputs_4,
    initial_outputs_5,
    initial_outputs_6,
    initial_outputs_7,
    initial_outputs_8,
]


# ======================
# MERGE ALL DATA
# ======================

X_final = []
y_final = []

for i in range(8):

    X_combined = np.vstack([
        initial_inputs[i],
        iter_inputs[i]
    ])

    y_combined = np.concatenate([
        initial_outputs[i],
        iter_outputs[i]
    ])

    X_final.append(X_combined)
    y_final.append(y_combined)

    print(f"Function {i+1}: {X_combined.shape}, {y_combined.shape}")


# ======================
# READY FOR MODEL
# ======================

# Example usage:
# for i in range(8):
#     q = choose_next_point_v7(X_final[i], y_final[i])
#     print(f"Function {i+1}: " + "-".join([f"{v:.6f}" for v in q]))
```

```python
import numpy as np
import tensorflow as tf
import random

def build_model(dim, hidden, lr):
    model = tf.keras.Sequential([
        tf.keras.Input(shape=(dim,)),
        tf.keras.layers.Dense(hidden, activation='relu'),
        tf.keras.layers.Dense(hidden, activation='relu'),
        tf.keras.layers.Dense(1)
    ])
    model.compile(
        optimizer=tf.keras.optimizers.Adam(learning_rate=lr),
        loss='mse'
    )
    return model


def tune_and_train(X, y_norm, dim):

    configs = []
    
    for _ in range(6):  # RANDOM SEARCH
        hidden = random.choice([8, 16, 32])
        lr = random.choice([0.001, 0.003, 0.01])
        epochs = random.choice([60, 100, 140])

        model = build_model(dim, hidden, lr)
        history = model.fit(X, y_norm, epochs=epochs, verbose=0)

        final_loss = history.history['loss'][-1]
        configs.append((model, final_loss))

    # pick best 3 models (ensemble)
    configs.sort(key=lambda x: x[1])
    best_models = [m for m, _ in configs[:3]]

    return best_models


def choose_next_point_v7(X, y):

    dim = X.shape[1]

    # normalize
    y_norm = (y - y.mean()) / (y.std() + 1e-8)

    # tuned ensemble
    models = tune_and_train(X, y_norm, dim)

    best_x = X[np.argmax(y)]

    candidates = []

    # (A) local search
    for _ in range(6):
        noise = np.random.normal(0, 0.04, dim)
        candidates.append(np.clip(best_x + noise, 0, 1))

    # (B) gradient refinement (more stable now)
    for m in models:
        x = tf.Variable(best_x.reshape(1, -1), dtype=tf.float32)

        for _ in range(20):
            with tf.GradientTape() as tape:
                pred = m(x)
            grads = tape.gradient(pred, x)

            grads = tf.clip_by_value(grads, -0.03, 0.03)
            x.assign_add(0.002 * grads)
            x.assign(tf.clip_by_value(x, 0, 1))

        candidates.append(x.numpy().flatten())

    # (C) exploration (slightly reduced)
    for _ in range(2):
        candidates.append(np.random.uniform(0, 1, dim))

    # scoring
    best_score = -np.inf
    best_candidate = None

    for x in candidates:
        preds = np.array([
            m.predict(x.reshape(1, -1), verbose=0)[0, 0]
            for m in models
        ])

        mu = preds.mean()
        sigma = preds.std()

        score = mu + 0.15 * sigma

        if score > best_score:
            best_score = score
            best_candidate = x

    return best_candidate
```

```python
# ======================
# GENERATE QUERIES
# ======================

queries = []

for i in range(8):
    X = X_final[i]
    y = y_final[i]

    q = choose_next_point_v7(X, y)
    queries.append(q)

# ======================
# FORMAT OUTPUT (IMPORTANT)
# ======================

for i, q in enumerate(queries):

    formatted = "-".join([f"{x:.6f}" for x in q])

    print(f"Function {i+1}: {formatted}")
```
